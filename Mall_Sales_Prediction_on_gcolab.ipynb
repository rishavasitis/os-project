# Importing Dataset through URL

```python
## Url of dataset
url = 'https://github.com/edyoda/data-science-complete-tutorial/blob/master/Data/Shopping_Revenue.csv'
```

```python
## converting the url to raw url for reading
url = url.replace('github.com','raw.githubusercontent.com').replace('/blob/','/')
```

```python
## Importing required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

%matplotlib inline
```

```python
## reading csv data from url
df1=pd.read_csv(url)
```

```python
## looking at the data
df1
```

```python
## checking info of the data
df1.info()
```

```python
## checking for null values
df1.isnull().sum()
```

#### as we can see P6 and P7 have null values so i will drop those rows

```python
df1.dropna(inplace=True)
df1
```

```python
## statical data
df1.describe()
```

# Visualization

```python
## city wise revenue
px.bar(df1.groupby('City').sum().sort_values('revenue'),y='revenue',title='City wise revenue',color='revenue')
```

```python
## type wise revenue
px.bar(df1,x='Type',y='revenue',color='Type',title='Type wise revenue')
```

```python
## city group wise revenue
px.histogram(df1,x='City Group',y='revenue',color='City Group',title='City Group wise revenue')
```

```python
## Converting City, City Group and Type column to categorical columns and encoding them (will be helpful in tree model)
df1['City'] = df1['City'].astype('category')
df1['City Group'] = df1['City Group'].astype('category')
df1['Type'] = df1['Type'].astype('category')
```

```python
df1['City'] = df1['City'].cat.codes
df1['City Group'] = df1['City Group'].cat.codes
df1['Type'] = df1['Type'].cat.codes
```

```python
## dropping id and date column
df1.drop(['Id','Open Date'],axis=1,inplace=True)
```

# Feature Selection

```python
## using heatmap for corelation visuals
plt.figure(figsize=(30,30))
sns.heatmap(df1.corr(),annot=True)
```

# Lets create Model

```python
## seperating dependent and independent values
X=df1.drop(['revenue'],axis=1)
Y=df1['revenue']
```

```python
## Splitting the data into training and testing
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=5)
```

## Decision tree regressor

```python
dr=DecisionTreeRegressor(random_state=5)
```

```python
## fitting the model
dr.fit(X_train,y_train)
```

```python
## prediction
y_pred=dr.predict(X_test)
```

```python
## to get scores
print('Scores : ')
print(' ')
print('\tTraining Score:', dr.score(X_train,y_train))
print('\tTesting Score:', dr.score(X_test,y_test))
print('\tR2 score: ',r2_score(y_test,y_pred))
print('\tRMSE score: ',np.sqrt(mean_squared_error(y_test,y_pred)))
```
